```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))

def build_plot(y):
    plt.plot(y)
    plt.grid(True)
    plt.show()
```

```python
# ex 1

x = np.array([1, 10, 1000])
numerator = 1/(np.power(np.e, np.sin(x)+1))
denominator = 5/4 + 1/x**15
y = np.log(numerator/denominator) / np.log(1+x**2)

print(y)
```

```python
# ex 2

x = np.arange(-4.0, 4.01, 0.01)
y = x**2-x-6

plt.plot(x, y, label=r'$f_1(x)=x**2-x-6$')
plt.plot(x, 0*x, label=r'$f_2(x)=0$')
plt.legend(loc='best', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# ex 3
x = np.arange(-100.0, 100.01, 0.01)

arg = (1 + x**2)*np.exp([-abs(x)/10])
base = 1 + np.tan([1/(1+np.power(np.sin(x), 2))])

y = np.array(np.log(arg)/np.log(base))
build_plot(y[0])
```

```python
# ex 4
input_function = input()
x = np.arange(-100.0, 100.01, 0.01)

def eval_input(inp):
    return eval(inp)

with plt.xkcd():
    try:
        y = eval_input(input_function)
        if len(y) == 1:
            y = y[0]
        plt.title('График введенной функции')
        build_plot(y)
    except Exception as e:
        print('Cannot build function graph,\n Eror: {}'.format(e))
```

```python
# ex 5

x = np.array([1, 2, 3, 4, 5])
y = np.array([0.99, 0.49, 0.35, 0.253, 0.18])
xarr = np.arange(x[0], x[-1], 0.01)
p2, v2 = np.polyfit(x, y, deg=2, cov=True)
p1, v1 = np.polyfit(x, y, deg=1, cov=True)

plt.subplot(121)
plt.errorbar(x, y, 
             xerr=np.mean([abs(v2[i][0]) for i in range(len(v2))]), 
             yerr=np.mean([abs(v2[i][1]) for i in range(len(v2))])
            )
plt.grid(True)
plt.plot(xarr, np.poly1d(p2)(xarr))

plt.subplot(122)
plt.errorbar(x, y, 
             xerr=np.mean([abs(v1[i][0]) for i in range(len(v1))]),
             yerr=np.mean([abs(v1[i][1]) for i in range(len(v1))])
            )
plt.plot(xarr, np.poly1d(p1)(xarr))
plt.grid(True)
plt.show()
```
